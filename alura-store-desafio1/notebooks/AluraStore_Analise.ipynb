```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

```python
faturamento = {
    "Loja 1": loja1["Preço"].sum(),
    "Loja 2": loja2["Preço"].sum(),
    "Loja 3": loja3["Preço"].sum(),
    "Loja 4": loja4["Preço"].sum()
}
print("Faturamento por loja:", faturamento)
```

```python
print("Loja 1 - Vendas por categoria:")
print(loja1["Categoria do Produto"].value_counts())

print("Loja 2 - Vendas por categoria:")
print(loja2["Categoria do Produto"].value_counts())
```

```python
avaliacoes = {
    "Loja 1": loja1["Avaliação da compra"].mean(),
    "Loja 2": loja2["Avaliação da compra"].mean(),
    "Loja 3": loja3["Avaliação da compra"].mean(),
    "Loja 4": loja4["Avaliação da compra"].mean()
}
print("Média de avaliações:", avaliacoes)
```

```python
print("Loja 1 - Mais vendidos:")
print(loja1["Produto"].value_counts().head(5))

print("Loja 1 - Menos vendidos:")
print(loja1["Produto"].value_counts().tail(5))
```

```python
frete = {
    "Loja 1": loja1["Frete"].mean(),
    "Loja 2": loja2["Frete"].mean(),
    "Loja 3": loja3["Frete"].mean(),
    "Loja 4": loja4["Frete"].mean()
}
print("Frete médio por loja:", frete)
```

```python
plt.bar(faturamento.keys(), faturamento.values(), color=['#4caf50','#2196f3','#ff9800','#f44336'])
plt.title("Faturamento por Loja")
plt.ylabel("R$ Total")
plt.show()
```

```python
loja1["Categoria do Produto"].value_counts().plot(kind="pie", autopct="%1.1f%%", figsize=(6,6))
plt.title("Distribuição de categorias - Loja 1")
plt.show()
```

```python
dados_frete = [loja1["Frete"], loja2["Frete"], loja3["Frete"], loja4["Frete"]]
plt.boxplot(dados_frete, labels=["Loja 1","Loja 2","Loja 3","Loja 4"])
plt.title("Comparação do frete médio entre lojas")
plt.ylabel("Valor do frete")
plt.show()
```

```python
import os
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

faturamento = {
    "Loja 1": loja1["Preço"].sum(),
    "Loja 2": loja2["Preço"].sum(),
    "Loja 3": loja3["Preço"].sum(),
    "Loja 4": loja4["Preço"].sum()
}

avaliacoes = {
    "Loja 1": loja1["Avaliação da compra"].mean(),
    "Loja 2": loja2["Avaliação da compra"].mean(),
    "Loja 3": loja3["Avaliação da compra"].mean(),
    "Loja 4": loja4["Avaliação da compra"].mean()
}

frete = {
    "Loja 1": loja1["Frete"].mean(),
    "Loja 2": loja2["Frete"].mean(),
    "Loja 3": loja3["Frete"].mean(),
    "Loja 4": loja4["Frete"].mean()
}

recomendacao = "Após análise, a Loja 4 apresenta o menor faturamento e deve ser vendida."

os.makedirs('reports', exist_ok=True)
with open("reports/relatorio_final.md", "w") as f:
    f.write("# 📊 Relatório Final - Alura Store\n\n")
    f.write("## Introdução\n")
    f.write("Este relatório apresenta a análise das quatro lojas da Alura Store, considerando faturamento, categorias vendidas, avaliações, produtos e frete.\n\n")

    f.write("## Resultados\n")
    f.write("- Faturamento por loja: " + str(faturamento) + "\n")
    f.write("- Média de avaliações: " + str(avaliacoes) + "\n")
    f.write("- Frete médio: " + str(frete) + "\n\n")

    f.write("## Conclusão\n")
    f.write(recomendacao)
```
